```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neural_network import MLPClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
# XOR dataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])

# Plotting
plt.figure(figsize=(8, 6))
for i, point in enumerate(X):
    if y[i] == 0:
        plt.scatter(point[0], point[1], color='red', marker='o', s=200)
    else:
        plt.scatter(point[0], point[1], color='yellow', marker='o', s=200)

plt.title("XOR Problem Visualization")
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid(alpha=0.8)
plt.show()
```

```python
# Train MLP
mlp = MLPClassifier(
    hidden_layer_sizes=(3,),  # (3,) indicates one hidden layer with 3 neurons.                      
    activation='tanh',        # 'tanh' (hyperbolic tangent) introduces non-linearity.                        
    max_iter=1000             # Sets the maximum number of iterations (epochs) 
)

mlp.fit(X, y)

# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 200),
                     np.linspace(-0.5, 1.5, 200))
Z = mlp.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
for i, point in enumerate(X):
    if y[i] == 0:
        plt.scatter(point[0], point[1], color='purple', marker='o', s=200)
    else:
        plt.scatter(point[0], point[1], color='yellow', marker='o', s=200)

plt.title("XOR Solved with MLP")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```
